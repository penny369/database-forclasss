```python
data=pd.read_csv('ckiptagger_data.csv')
```

```python
data.head(4)
```

```python
df=data.dropna(inplace=False)
df=df.reset_index(drop=True)
df
```

```python
TCRI=[]
for i in range(0,len(df)):
    TCRI.append(df.iloc[i,3][0])
```

```python
df=df.drop(columns=['TCRI(年/月)'])
len(TCRI)
```

```python
df.insert(3, 'TCRI', TCRI)
```

```python
words=[]
for word in df.content_tokenized.iloc[:6000]:
    a=word.split(" ")
    for b in a:
        words.append(b)
```

```python
words=pd.unique(words)
```

```python
mydata=[]
onematrix=[]
```

```python
import math
for a in df.iloc[0:6000,7]:
    for b in words:
        count = 0
        if (a.find(b)>=0):
            count += 1
        onematrix.append(count)
    mydata.append(onematrix)
    onematrix=[]
```

```python
data1=pd.DataFrame(mydata,columns=words)
```

```python
newdata=pd.concat([df.iloc[:6000],data1],axis=1)
newdata
```

```python
x=newdata.iloc[:,9:].values
y=newdata.iloc[:,5].values

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
pd.unique(newdata.iloc[:,5])
```

```python
list(le.classes_)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                              test_size=0.2,
                                              random_state=1,
                                              stratify=y)
```

```python
from sklearn.svm import SVC
svm=SVC(kernel="linear",probability=True)
```

```python
svm.fit(x_train,y_train)
```

```python
y_test_predict = svm.predict(x_test)
```

```python
y_test_predict[y_test_predict==4]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_test_predict))
print(classification_report(y_test,y_test_predict))
```
